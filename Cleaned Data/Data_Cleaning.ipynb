Data Cleaning

```python
df_copy=df.copy()           # Making Copy
```

```python
df_copy.columns

lst_of_columns=list(df_copy.columns)

print(lst_of_columns)
```

```python
for col in lst_of_columns:
  print(f"Column : {col}")
  print("")
  print(df_copy[col].unique())
  print("")
  print("----------------------------")
```

```python
# Handling "Category" feature
```

```python
df_copy["Category"].unique()
```

```python
df_copy["Category"]=df_copy["Category"].replace("1.9",np.nan)
```

```python
# Handling "Reviews" feature: Changing Data type
```

```python
df_copy["Reviews"].astype(int)
```

```python
df_copy["Reviews"].str.isnumeric().sum()
```

```python
df_copy[~df_copy['Reviews'].str.isnumeric()]
```

```python
df_copy=df_copy.drop(df_copy.index[10472])          ##Droping Value
```

```python
df_copy["Reviews"]=df_copy["Reviews"].astype(int)
```

```python
# Handling Features "Size","Installs","Price".

#  To handle :
#  1) "M" and "K" from Size
#  2) "+"  and  "," from Installs
#  3) "$" from Price
```

```python
obj_to_remove=["k","+",",","$"]
columns_for_obj=["Size","Installs","Price"]

for obj in obj_to_remove:
    for col in columns_for_obj:
        df_copy[col]=df_copy[col].str.replace(obj,"")
```

```python
df_copy["Size"]=df_copy["Size"].str.replace("M","000")
df_copy["Size"]=df_copy["Size"].replace("Varies with device",np.nan)
```

```python
df_copy["Installs"]=df_copy["Installs"].str.replace("Free","0")
```

```python
df_copy["Size"]=df_copy["Size"].astype(float)
df_copy["Installs"]=df_copy["Installs"].astype(int)
df_copy["Price"]=df_copy["Price"].astype(float)
```

```python
# Handling "Last Updated" feature
```

```python
df_copy["Last Updated"]
```

```python
df_copy['Last Updated']=pd.to_datetime(df_copy['Last Updated'])
```

```python
df_copy['Day']=df_copy['Last Updated'].dt.day
df_copy['Month']=df_copy['Last Updated'].dt.month
df_copy['Year']=df_copy['Last Updated'].dt.year
```

```python
df_copy.info()
```

```python
df_copy.to_csv('Cleaned Google Play Store File.csv', index=False)
```
